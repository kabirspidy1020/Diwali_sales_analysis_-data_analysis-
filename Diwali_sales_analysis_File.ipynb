**EXPLORATORY DATA ANALYSIS**

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

Above graph shows that females purchased were more as comparson to males

```python
sales_by_genders=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data= sales_by_genders)
```

From above graph most of the buyers are women and even the purcahasing power of women is more in comparison with men

```python
# age
ax=sns.countplot(data=df, x='Age Group', hue='Gender') #hue is used to  divide bar in genders of f and M
```

```python
# age
ax=sns.countplot(data=df, x='Age Group', hue='Gender') #hue is used to  divide bar in genders of f and M
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax=sns.countplot(data=df, x='Age Group')# without hue
```

```python
sales_by_age_group=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data= sales_by_age_group)
```

**From above chart we can say that most of the buyers are from age group 26-35 and most of the womens are interested in shoppings**

```python
#state wise
ax=sns.countplot(data=df, x='State', hue='Gender')
sns.set(rc={'figure.figsize':(25,5)})
```

```python
sales_by_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(x='State', y='Orders', data= sales_by_state)
```

From above graphs we can conclude that most people ordered from Uttar pradesh

```python
sales_by_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(x='State', y='Amount', data= sales_by_state)
```

Haryana took place of kerala, since haryana placed less orders than kerala but haryana ordered power is more i.e they ordered more amount or expensive amount items

```python
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for i in ax.containers:
  ax.bar_label(i)
```

```python
#product_category
ax=sns.countplot(data=df, x='Product_Category')
for i in ax.containers:
  ax.bar_label(i)
```

It is quite clear from above graph people are ordering more in Clothing and apparel, food and electronics and gadgets

```python
sales_by_Product_Category=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Product_Category', y='Amount', data= sales_by_Product_Category)
```

amount= food> Clothings and apparel> electronics and gadgets

orders=  Clothings and apparel > food>  electronics and gadgets

```python
#printing 10 largest selling products on the basis of product id
Top_10_product=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='Product_ID', y='Orders', data= Top_10_product)
```

**From exploratory data analysis we can see that womens of 26-35 age belonging to it sectors, healthcare and aviation are ordering more in category of Foods, Clothings and apparels, electronics and gadgets**
